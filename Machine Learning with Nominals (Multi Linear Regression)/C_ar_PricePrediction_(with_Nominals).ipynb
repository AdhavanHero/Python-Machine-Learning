```python
from google.colab import files
uploaded = files.upload()
```

```python
import numpy as np 
import pandas as pd
from sklearn import linear_model

import math
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['carprices.csv']))
```

```python
df
```

```python
dummies = pd.get_dummies(df,['Car Model'])
```

```python
dummies = dummies.drop(['Sell Price($)','Mileage','Age(yrs)'],axis='columns')
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['Car Model','Car Model_BMW X5'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = final.drop('Sell Price($)',axis='columns')
```

```python
x
```

```python
y = final['Sell Price($)']
```

```python
model.fit(x,y)
```

```python
model.predict([[59000,5,0,1]])
```

```python
model.score(x,y)
```
